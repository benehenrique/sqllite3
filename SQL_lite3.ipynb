```python
import sqlite3
from datetime import date, timedelta
import random

connection=sqlite3.connect('icd.db') #conexão
c = connection.cursor() #cursor
c.execute('CREATE TABLE IF NOT EXISTS tabela(Conta TEXT, Data TEXT, Valor INTEGER, x INTEGER PRIMARY KEY)') #td q for sql por em maiúsculo
c.execute('CREATE TABLE IF NOT EXISTS tipo(A INTEGER, B INTEGER, C INTEGER, x INTEGER, Data TEXT PRIMARY KEY, FOREIGN KEY (x) REFERENCES tabela(x))')
```

```python
Data=[]
x=[]

for i in range(30):
    x.append(i)
    Data=((date.today()+timedelta(days=i)).isoformat())
    Valor=random.randrange(-5000, 5000)
    A=Valor/2
    B=Valor/4
    C=Valor/6
    if Valor > 0:
        conta = 'Positivo'
    elif Valor < 0:
        conta = 'Negativo'
    else:
        conta = 'Neutro'
    c.execute("INSERT INTO tabela(Conta, Data, Valor, x)VALUES (?,?,?,?)",(conta, Data, Valor, x[i]))
    c.execute("INSERT INTO tipo(A,B,C,x,Data)VALUES(?,?,?,?,?)",(A,B,C,x[i],Data))
    connection.commit()
```

```python
c.execute('SELECT * FROM tipo')
for data in c.fetchall():
    print(data)
#c.close()
#connection.close()
```
